```python
import torch
import torch.optim 
from TorchProteinLibrary import *
import Bio
import numpy as np
import _Volume
import matplotlib.pylab as plt
import matplotlib.animation
import numpy as np
import mpl_toolkits.mplot3d.axes3d as p3
from functools import partial
from torch import nn as nn
from torch.nn import functional as F
from pyrosetta import *;pyrosetta.init()

class Energy(nn.Module):
    """
    We can implement our own custom autograd Functions by subclassing
    torch.autograd.Function and implementing the forward and backward passes
    which operate on Tensors.
    """
    def __init__(self,seq_name,res_names,atom_names,num_atoms,up_ang):
        super(Energy, self).__init__()
#         print(seq_name)
        self.pose=pose_from_sequence(seq_name)
        self.seq_name=seq_name
        self.res_names=res_names
        self.atom_names=atom_names
        self.num_atoms=num_atoms
        self.nrg=energy_func.apply

        self.up_ang=up_ang
        self.define_energy()
        self.gen_dof_list()
        self.create_population_list()
        ## Set up the scoring function and the movers for the pose
        
    
    def forward(self, new_angles, new_coordinates):
        """
        In the forward pass we receive a Tensor containing the input and return
        a Tensor containing the output. ctx is a context object that can be used
        to stash information for backward computation. You can cache arbitrary
        objects for use in the backward pass using the ctx.save_for_backward method.
        """
#         print(new_coordinates)
        ##Initialize our pose with correct positions
        self.update_pose(new_angles,new_coordinates)
        self.min_map.copy_dofs_from_pose( self.pose, self.dofs );
        self.shape = new_angles.shape
        self.dtype= new_angles.type
        return self.nrg(new_angles,self.multifunc,self.dofs,self.dof_locs)


    
    def gen_dof_list(self):
        self.dof_locs=[]
        ##
        ###  Find which angles dofs correspond to
        for i in range(len(self.dofs)):
            rsd = self.min_map.dof_nodes().front().torsion_id().rsd()
            ang = self.min_map.dof_nodes().front().torsion_id().torsion()
            if(rosetta.core.id.to_string(self.min_map.dof_nodes().front().torsion_id().type())=='SC'):
                ang+=3
            self.min_map.dof_nodes().push_back(self.min_map.dof_nodes().front())
            self.min_map.dof_nodes().pop_front()
            self.min_map.dof_nodes().resize(len(self.dofs))
            self.dof_locs.append((ang,rsd))
#         print(self.dof_locs)
        
    
    def create_population_list(self):
        res_num=0
        self.pop_list=[]
        cur_res_list=[]
        for i,at in enumerate(self.atom_names[0].detach()):
            atom_name=''
            for j in at.numpy():
                if(j!=0):atom_name+=chr(j)
                else:break
            if(atom_name=='N'):
                if(res_num!=0):
                    self.pop_list.append(cur_res_list)
                res_num+=1
                cur_res_list=list(range(1,self.pose.residue(res_num).natoms()+1))
#                 print(cur_res_list)
            atom=self.pose.residue(res_num).atom_index(atom_name)
            if(atom in cur_res_list):
                cur_res_list.remove(atom)       
        self.pop_list.append(cur_res_list)

        
    
    def update_pose(self,new_angles,new_coordinates):
        """
        We are first going to populate the pose with the correct positions and angles
        """

        ###
        ### Setting all coordinates from our Torch implementation to the PyRosetta one

        res_num=0
        at_ids= rosetta.utility.vector1_core_id_AtomID()
        xyzs=pyrosetta.rosetta.utility.vector1_numeric_xyzVector_double_t()
        for i,at in enumerate(self.atom_names[0].detach()):
            atom_name=''
            for j in at.numpy():
                if(j!=0):atom_name+=chr(j)
                else:break
#             print(atom_name)
            if(atom_name=='N'):
                res_num+=1
#             print(atom_name)
            atom=self.pose.residue(res_num).atom_index(atom_name)
            at_id = AtomID(atom,res_num)
#             print(atom_name,at_id)
#             import pdb; pdb.set_trace()
            xyz=new_coordinates.view(-1,3)[i,:].detach().numpy()
            xyzs.append(rosetta.numeric.xyzVector_double_t(xyz[0],xyz[1],xyz[2]))
            at_ids.append(at_id)
#             if(atom_name=='C'):
#                 print(xyz,atom,res_num)
        self.pose.conformation().batch_set_xyz(at_ids, xyzs)        
        
        
        """
        We must now also move the hydrogen atoms (and others) to the correct positions,
        as they dont get moved automatically for some reason
        """

        for res in range(1,len(self.pose.sequence())+1):
            for atoms in self.pop_list[res-1]:#range(1,self.pose.residue(res).natoms()+1): 
#                 if(self.pose.residue(res).atom_is_hydrogen(atoms) or 'V' in self.pose.residue(res).atom_name(atoms) or 'OXT' in self.pose.residue(res).atom_name(atoms)):
                at_id = AtomID(atoms,res)
                self.pose.conformation().set_xyz(at_id,self.pose.residue(res).build_atom_ideal(atoms,self.pose.conformation()))
        
        
        angles_in_deg=new_angles*180/np.pi
        for res in range(len(self.pose.sequence())):
            ###
            ### Setting angles
            self.pose.set_phi(res+1,angles_in_deg[0,0,res])
            self.pose.set_psi(res+1,angles_in_deg[0,1,res])
            self.pose.set_omega(res+1,angles_in_deg[0,2,res])
            for chis in range(len(self.pose.residue(res+1).chi())):
                self.pose.set_chi(chis+1,res+1,angles_in_deg[0,3+chis,res])

        
        
            
#         self.pose.conformation().update_actcoords()
    def define_energy(self):
        self.scorefxn = get_score_function()
        self.move_map = rosetta.core.kinematics.MoveMap()  # Adjust heavily to pull out just the degrees of freedom you want
        start_score = self.scorefxn(self.pose)
#         print(start_score)
        self.min_map = rosetta.core.optimization.MinimizerMap() # Need to change this to CartesianMinimizerMap for cartmin
        self.move_map.set_bb(True)
        self.move_map.set_chi(True)
        self.min_map.setup( self.pose, self.move_map )
        self.scorefxn.setup_for_minimizing( self.pose, self.min_map )
        self.multifunc = rosetta.core.optimization.AtomTreeMultifunc( self.pose, self.min_map, self.scorefxn ) 
        self.dofs = Vector1( [0.0]* (self.min_map.nangles()))

class energy_func(torch.autograd.Function):
    """
    We can implement our own custom autograd Functions by subclassing
    torch.autograd.Function and implementing the forward and backward passes
    which operate on Tensors.
    """

    @staticmethod
    def forward(ctx, input, multifunc,dofs,dof_locs):
        """
        In the forward pass we receive a Tensor containing the input and return
        a Tensor containing the output. ctx is a context object that can be used
        to stash information for backward computation. You can cache arbitrary
        objects for use in the backward pass using the ctx.save_for_backward method.
        """
        ctx.multifunc=multifunc
        ctx.dofs=dofs
        ctx.dof_locs=dof_locs
        ctx.shape = input.shape
        ctx.dtype=input.type()
        return torch.Tensor([multifunc( dofs )])

    @staticmethod
    def backward(ctx, grad_output):
        """
        In the backward pass we receive a Tensor containing the gradient of the loss
        with respect to the output, and we need to compute the gradient of the loss
        with respect to the input.
        """
#         multifunc,dofs,dof_list,shape,dtype = ctx.saved_tensors
        varis = Vector1( list( ctx.dofs ) )
        dE_dvars = Vector1( list( ctx.dofs ) )

        ctx.multifunc.dfunc( varis, dE_dvars )
        grad_input=torch.zeros(ctx.shape).type(ctx.dtype)
#         print(dE_dvars,varis)
        for dof,dof_grad in enumerate(dE_dvars):
            grad_input[0,ctx.dof_locs[dof][0]-1,ctx.dof_locs[dof][1]-1]=dof_grad
        return grad_input.type_as(grad_output)*grad_output*np.pi/180.0, None, None, None

a2c = FullAtomModel.Angles2Coords()
rmsd = RMSD.Coords2RMSD()
vc=Volume.VolumeConvolution()
tc2v=Volume.TypedCoords2Volume(box_size=120)

#Coords transforms
c2c = FullAtomModel.Coords2Center()
translate = FullAtomModel.CoordsTranslate()
rotate = FullAtomModel.CoordsRotate()

p2c = FullAtomModel.PDB2CoordsUnordered()
c2tc = FullAtomModel.Coords2TypedCoords()
```

```python
def seq_from_res_name(res_names_dst,atom_names_dst):
    is0C = torch.eq(atom_names_dst[:,:,0], 67).squeeze()
    is1A = torch.eq(atom_names_dst[:,:,1], 65).squeeze()
    is20 = torch.eq(atom_names_dst[:,:,2], 0).squeeze()
    isCA = is0C*is1A*is20
    out=[]
    out_str=''
    dicti={ 'GLY':'G',
            'ALA':'A',
            'LEU':'L',
            'MET':'M',
            'PHE':'F',
            'TRP':'W',
            'LYS':'K',
            'GLN':'Q',
            'SER':'S',
            'PRO':'P',
            'VAL':'V',
            'ILE':'I',
            'CYS':'C',
            'TYR':'Y',
            'HIS':'H',
            'ARG':'R',
            'ASN':'N',
            'ASP':'D',
            'GLU':'E',
            'THR':'T'}
    for i,strin in enumerate(res_names_dst.squeeze().numpy()):
        out_str=''
        for char in (strin):
            if char!=0:
                out_str+=(chr(char))
#         print(out_str)
                
        if isCA[i]:
            if(out_str in dicti):
                out_str=dicti[out_str]
            
            out.append(out_str)
#         if(i==109):
#             print(out_str)
#             break
#         print(i)
    return [''.join(out)]
```

```python
## Rotates B_true into B_1

def rotation_mat(B_true_1,B_1):
#     print(B_true_1)
    B_true_2=B_true_1/torch.norm(B_true_1,2)
    B_2=B_1/torch.norm(B_1,2)
    cosin=torch.dot(B_true_2.squeeze(),B_2.squeeze())
    sin=(torch.cross(B_2.squeeze(),B_true_2.squeeze()))
#     print(cosin,sin)
    rot_v=torch.tensor([[0,-sin[2],sin[1]],[sin[2],0,-sin[0]],[-sin[1],sin[0],0]])
    rotation=torch.eye(3, 3, dtype=torch.float).unsqueeze(dim=0) + rot_v + torch.mm(rot_v,rot_v)*1/(1+cosin)
#     print(rotation.shlape)

#     print(torch.mv(rotation.squeeze(),B.squeeze()))
#     print(B_true)
#     print(torch.mm(rotation.squeeze(),(rotation.squeeze().t())))
    return rotation.numpy()
```
